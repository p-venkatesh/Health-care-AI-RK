```python
import matplotlib.image as mpimg
import random 
from skimage import exposure 
```

```python
from skimage.util import random_noise
from skimage import transform 
from cv2 import resize
```

```python
img = mpimg.imread('/Users/p.venkatesh/Desktop/Screenshot.jpg')
```

```python
plt.imshow(img)
img_rescale=resize(img, (400, 400))
plt.imshow(img_rescale)
```

```python
#Horizontal flip
horizontal = np.fliplr(img_rescale)
plt.imshow(horizontal)
```

```python
#Vertical flip
vertical = np.flipud(img_rescale)
plt.imshow(vertical)
```

```python
#rotate
from skimage import transform 
transf = transform.rotate(img_rescale, random.uniform(-20, 20))
plt.imshow(transf)
```

```python
#random noise
img_noise = random_noise(img_rescale, mode = 's&p', clip = True)
plt.imshow(img_noise)
#mpimg.imsave("noise_img_dog", img_noise)
```
